## Model Loss:
We have that the model loss is:

$$ L = \frac{1}{n}\sum_{i=1}^{n}(Model(X_{i})-Y_{i}))^2 $$

For us to use gradient descent, let's compute the derivative of the loss with respect to the Weights matrix.

$$Model(X_i) = (mx_i+b-y_i)^2$$

And, for the derivative with respect to m:

$$ \frac{\delta }{\delta m} (mx_i+b-y_i)^2 = $$
$$ = 2(mx_i+b-y_i)*\frac{\delta }{\delta m}(mx_i+b-y_i) =  $$
$$ =  2(mx_i+b-y_i) * x_i  $$


For the derivative with respect to b:

$$ \frac{\delta }{\delta b} (mx_i+b-y_i)^2 = $$
$$ = 2(mx_i+b-y_i)*\frac{\delta }{\delta b}(mx_i+b-y_i) =  $$
$$ =  2(mx_i+b-y_i)  $$

Our update function is going to be:

$$W_j = W_j - learning_{rate} * \frac{\delta L}{\delta W_j}$$

```python
import numpy as np

class SimpleLinearRegression:
    def __init__(self, learning_rate = 1e-3, n_steps = 1000):
        self.learning_rate = learning_rate
        self.n_steps = n_steps
        
    def get_W(self):
        return self.W
    
    def compute_loss(self, X, Y):
        X_to_use = np.c_[np.ones(X.shape[0]), X]
        inner = np.dot(X_to_use, self.W)
        loss = (2/X_to_use.shape[0]) * sum(pow( inner - Y, 2))
        return loss
    
    def fit(self, X, Y):
        X_train = np.c_[np.ones(X.shape[0]), X] # Bias term added!
        self.W = np.random.rand((X_train.shape[1])) # Random initialization for the model weights
        
        for i in range(self.n_steps):
            self.W = self.W - self.learning_rate * self.calcGradient(X_train, Y)
            
    
    def predict(self, X):
        X_pred = np.c_[np.ones(X.shape[0]), X]
        return np.dot(X_pred, self.W)
    
    def calcGradient(self, X, Y):
        return 2/X.shape[0] * np.dot(X.T, (np.dot(X, self.W) - Y))
```

```python
model = SimpleLinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(np.asarray([1]))
print(y_pred)
print(model.compute_loss(X_train, y_train))

```

```python
X_new = X_train.tolist()
y_new = y_train.tolist()

X_new.append(1)
y_new.append(model.predict(np.asarray([1])))

W = model.get_W()

x = np.linspace(-4,4,100)
y = W[0] + x*W[1]

figure(figsize=(4, 4))

pyplot.grid()
pyplot.plot(x, y, '-')
pyplot.scatter(X_new, y_new)
```
